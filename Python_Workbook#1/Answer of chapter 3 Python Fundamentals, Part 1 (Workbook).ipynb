```python
#Write your salary computation to give an employee 1.5 times the hourly rate for hours worked above 40 hours. Example:

#Enter Hours: 45
#Enter Rate: 10
#Pay: 475.0

hourworked = float(input("Enter Hours: "))
hourrate = float(input("Enter Rate: "))

if hourworked > 40:
  regular_hourworked = 40
  overtimehour = hourworked -  regular_hourworked
else :
  regular_hourworked = hourworked
  overtimehour = 0

regularpay = regular_hourworked * hourrate
overtimepay = overtimehour * 1.5 * hourrate

total_sal = regularpay + overtimepay

print("Total pay: ", total_sal)
```

```python
#Write a program to prompt for a score between 0 and 100 and print a grade using the following table:

#Score           Grade
#[ 97, 100]      1.0
#[ 93,  97)        1.25
#[ 89,  93)        1.5
#[ 85,  89)        1.75
#[ 81,  85)        2.0
#[ 77,  81)        2.25
#[ 74,  77)        2.5
#[ 0,   74)        3.0

score = float(input("Enter your score: "))

if score >=97 and score <= 100:
    grade = 1.0
elif score >= 93 and score <= 97:
    grade = 1.25
elif score >= 89 and score <= 93:
    grade = 1.5
elif score >= 85 and score <= 89:
    grade = 1.75
elif score >= 81 and score <= 85:
    grade = 2.0
elif score >= 77 and score <= 81:
    grade = 2.25
elif score >= 74 and score <= 77:
    grade = 2.5
elif score >= 0 and score <= 74:
    grade = 3.0

print("Your Grade is:", grade)


```
