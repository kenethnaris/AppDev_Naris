Exercise num 3:

```python
#Assume that we execute the following assignment statements:
#width = 17
#height = 12.0
#For each of the following expressions, write the value of the expression and the type (of the value of the expression).
#width//2
#width/2.0
#height/3
#1 + 2 * 5


width = 17
height = 12.0

#When we store the width//2 to "a" and use the type() function and call it inside the print() function, it will prompt the data type "int"
#because the answer of 17 // 2 = 8 in which these "//" operator called floor division leaves the answer a whole number and disregards the remainder.
a = width//2
type(a)


#When we store the width/2.0 to "b" and use the type() function and call it inside the print() function, it will prompt the data type "float"
#because the answer of 17 / 2.0 = 8.5 in which these "/" operator called division leaves the answer as float or a decimal number.
b = width/2.0
type(b)


#When we store the height/3 to "c" and use the type() function and call it inside the print() function, it will prompt the data type "float"
#even though the answer of 12.0 / 3 = 4 because it will prioritize the float data type which is the given "height" above.
c = height/3
type(c)


#When we store the 1 + 2 *5 to "d" and use the type() function and call it inside the print() function, it will prompt the data type "int"
#because the answer of 1 + 2 *5 = 11 which is an integer.
d = 1 + 2 *5
type(d)

print(f"{type(a)},{type(b)},{type(c)},{type(d)}")



```
