```python
#Rewrite your pay computation with time-and-a-half for overtime and create a function called compute pay which takes two parameters (hours and rate).

#Enter Hours: 45
#Enter Rate: 10
#Pay: 475.0
def compute(hours, rate):
    if hours > 40:
        regular_hourworked = 40
        overtimehour = hours -  regular_hourworked
    else :
        regular_hourworked = hours
        overtimehour = 0

    regularpay = regular_hourworked * rate
    overtimepay = overtimehour * 1.5 * rate

    total_sal = regularpay + overtimepay

    return total_sal


hours = float(input("Enter Hours: "))
rate = float(input("Enter Rate: "))


print("Total pay: ", compute(hours, rate))
```

```python
#Write a function called computegrade that takes a score as its parameter and returns a grade as a string using the following table:

#Score           Grade
#[ 97, 100]      1.0
#[ 93,  97)        1.25
#[ 89,  93)        1.5
#[ 85,  89)        1.75
#[ 81,  85)        2.0
#[ 77,  81)        2.25
#[ 74,  77)        2.5
#[ 0,   74)        3.0

#Run the program repeatedly to test the various different values for input.


def computegrade(score):
  if score >= 97 and score <= 100:
    return 1.0
  elif score >= 93 and score <=97:
    return 1.25
  elif score >= 89 and score <=93:
    return 1.5
  elif score >= 85 and score <=89:
    return 1.75
  elif score >= 81 and score <=85:
    return 2.0
  elif score >= 77 and score <=81:
    return 2.25
  elif score >= 74 and score <=77:
    return 2.5
  elif score >=0 and score <=74:
    return 3.0

score = float(input("Enter a score between 0 to 100: "))
grade = computegrade(score)
print(f"Your Grade is: {grade}")
```
