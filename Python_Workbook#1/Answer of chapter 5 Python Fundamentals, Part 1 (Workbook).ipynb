The "is" keyword in Python is used for identity comparison, not assignment.
If you want to check if two variables point to the same object in memory, you can use the "is" keyword.

For lists and other compound objects, each object is created separately and will not be reused, even if the elements of the list are the same. That's why when you compare two lists with the same elements using the 'is' keyword, it returns False.

```python
#example
a = 2
b = 2.0
c = a

a = [2]
c = a

print(a is c)
```

```python
#example when the elements inside the list are just the same
a = 2
b = 2.0
c = a

a = [2]
b = [2]
c = a

print(a is b)
```

== is an operator that checks if the values of the two variables are equal, regardless of whether they are the same object or not. If you want to check if two variables have the same value, you can use the == operator.

```python
#example of == using the given variables

a = 2
b = 2.0
c = a

print(a == b)
```

Even though a is an integer and b is a floating-point number, they have the same value of 2. Python automatically converts a to a floating-point number to do the comparison, which is why the == operator returns True.

Write a program which repeatedly reads numbers until the user enters "done". Once "done" is entered, print out the total, count, and average of the numbers. If the user enters anything other than a number, detect their mistake using try and except and print an error message and skip to the next number.

Enter a number: 4
Enter a number: 5
Enter a number: bad data
Invalid input
Enter a number: 7
Enter a number: done
16 3 5.333333333333333

```python
num = 0
tot = 0.0

while True:
    number = input("Enter a number: ")
    if number == 'done':
        break
    try:
        num1 = float(number)
    except:
        print("Invalid input")
        continue
    num = num + 1
    tot = tot + num1

print("Total: ", tot)
print("Count: ", num)
print("Average: ", tot/num)
```

Write another program that prompts for a list of numbers as above and at the end prints out both the maximum and minimum of the numbers instead of the average.

```python
largest = None
smallest = None
count = 0
total = 0

while True:
    try:
        number = input("Enter a number: ")
        if number == "done":
            break
        number = int(number)
        count +=1
        total =number + total
        if largest is None or largest < number:
            largest = number
        elif smallest is None or smallest > number:
            smallest = number
    except:
        print("Invalid input")

print("The maximum number is: {}".format(largest))
print("The minimum number is: {}".format(smallest))
print("The total number is: {}".format(total))
print("The count number is: {}".format(count))
```
